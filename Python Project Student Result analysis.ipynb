# Now checking impact of parents education on students marks

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","WritingScore":"mean","ReadingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot=True)
plt.title("Impact of Parents Education on Students marks")
plt.show()
```

# Conclusion:Education of parents have a good impact on students marks 

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
```

```python
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1,annot=True)
plt.title("Impact of Parents Marital Status on Students Marks")
plt.show()
```
